```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
data = pd.read_csv('/content/Melbourne_housing_FULL.csv')
```

```python
data.head(5)
```

```python
data.nunique()
```

```python
data.shape
```

```python
df =  data [['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount','Distance', 'CouncilArea',
             'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']]
df.head(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
cols_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_fill_zero] = df[cols_fill_zero].fillna(0)
df.isnull().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df=pd.get_dummies(df,drop_first=True).astype(int)
df.head()
```

```python
x = df.drop('Price',axis=1)
y = df['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

Which is a Overfitting and we use Regularization techiques for this

1. Lasso regression - L1

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha = 50,max_iter=100,tol=0.1)
lasso.fit(x_train,y_train)
```

```python
lasso.score(x_test,y_test)
```

```python
lasso.score(x_train,y_train)
```

```python
print(lasso.coef_)
```

2. Ridge Regression - L2

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha = 50,max_iter=100,tol=0.1)
ridge.fit(x_train,y_train)
```

```python
ridge.score(x_test,y_test)
```

```python
ridge.score(x_train,y_train)
```

```python
print(ridge.coef_)
```

3. Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet
```

```python
el = ElasticNet(alpha = 1.0,l1_ratio=0.5)
el.fit(x_train,y_train)
```

```python
el.score(x_test,y_test)
```

```python
el.score(x_train,y_train)
```
